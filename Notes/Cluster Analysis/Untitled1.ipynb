```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a = np.random.randint(0, 100, 350)
b = np.random.randint(0, 100, 350)
```

```python
a
```

```python
b
```

```python
df = pd.DataFrame({'a':a, 'b':b})
```

```python
df
```

```python
sns.histplot(x=a, data=df, bins=20)
plt.show()
```

```python
sns.histplot(x=b, data=df, bins=20)
plt.show()
```

```python
sns.swarmplot(y=a, data=df)
```

```python
sns.swarmplot(y=b, data=df)
```
