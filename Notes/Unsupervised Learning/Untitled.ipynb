# Hierarchical clustering of the grain data

In the video, you learned that the SciPy linkage() function performs hierarchical clustering on an array of samples. Use the linkage() function to obtain a hierarchical clustering of the grain samples, and use dendrogram() to visualize the result. A sample of the grain measurements is provided in the array samples, while the variety of each grain sample is given by the list varieties. 

```python
# Perform the necessary imports
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Calculate the linkage: mergings
mergings = linkage(samples, method='complete')

# Plot the dendrogram, using varieties as labels
dendrogram(mergings,
           labels=varieties,
           leaf_rotation=90,
           leaf_font_size=6,
)
plt.show()
```

# Hierarchies of stocks

In chapter 1, you used k-means clustering to cluster companies according to their stock price movements. Now, you'll perform hierarchical clustering of the companies. You are given a NumPy array of price movements movements, where the rows correspond to companies, and a list of the company names companies. SciPy hierarchical clustering doesn't fit into a sklearn pipeline, so you'll need to use the normalize() function from sklearn.preprocessing instead of Normalizer.

linkage and dendrogram have already been imported from scipy.cluster.hierarchy, and PyPlot has been imported as plt. 

```python
stocks = pd.read_csv('data/stock_price.csv')
stocks.head(1)
movements = stocks.drop('Unnamed: 0', axis=1).values
companies = stocks['Unnamed: 0'].values
```

```python
companies
```

```python
# Import normalize
from sklearn.preprocessing import normalize

# Normalize the movements: normalized_movements
normalized_movements = normalize(movements)

# Calculate the linkage: mergings
mergings = linkage(normalized_movements, method='complete')

# Plot the dendrogram
plt.figure(figsize=(20, 15))
dendrogram(mergings, labels=companies, leaf_rotation=90, leaf_font_size=12)
plt.show()
```

# Cluster Labels in Hierachical clustering
* Not only a visualization tool
* Cluster labels at any intermediate stage can be recovered
    * can be used for calculations like cross tabulations
* we can specify a height on the dendrogram 
    * if we specify height of 15, it w
    * the y-axis on the dendrogram specifies the distance between merging clusters
    * the distance is computed with the linkage method
        * we used "complete" above
        * the distance betweens is max distance between their samples
        * different linkage methods yield different hierarchical clustering
* extracting cluster labels
    * fcluster()

```python
from scipy.cluster.hierarchy import linkage
mergings=linkage(movements, method='complete')
from scipy.cluster.hierarchy import fcluster

#pass result from the linkage function to the fcluster() function
# 15 denotes the height
labels = fcluster(mergings, 15, criterion='distance')
print(labels)
```

### Using linkage method='single' produces a very different diagram
* this is because single linkage uses the distance of the closest point to the next cluster, instead of the farthes, as is done with 'complete' linkage

# A t-SNE map of the stock market

t-SNE provides great visualizations when the individual samples can be labeled. In this exercise, you'll apply t-SNE to the company stock price data. # Finds patters in data and reexpresses in compressed form
* makes subsequent computations much easier
* however: the most important function of dimension reduction is to reduce a dataset to its bare bones
    * i.e. remove less informative noise features
    * getting rid of the noise makes prediction more feasible
    
### PCA
* takes place in 2 steps
    * decorrelation does not change the dimension of the data at all

* the fit method learns how to shift and rotate the samples but does not change them
* transform() applies the transformation the fit method learns
    * can be applied to new, unseen samples

# Dimension Reduction with PCA
* represent the same data using less features
* important for building ml pipelines
* We use PCA for this

* PCA features are in decreasing order of variance
* assumes the low variance features as noise, so in the graph above, FEATURES 1, 2  are noise
    * Tell the model how many features to keep
        * PCA(n_components=2)

```python
samples = iris.data
species = iris.target

from sklearn.decomposition import PCA
pca = PCA(n_components=2)

pca.fit(samples)

transformed = pca.transform(samples)
print(transformed.shape)
```

### We can now plot this

```python
import matplotlib.pyplot as plt
xs=transformed[:, 0]
ys=transformed[:, 1]
plt.figure(figsize=(20,15))
plt.scatter(xs, ys, c=species)
plt.show()
```

# Non-negative matrix factorization
### NMF, like PCA is a dimension reduction technique
* Unlike PCA, NMF models are interpretable
* this means easier to understand and much easier you to explain to others
* cannot be applied to every dataset however
    * it is required that the sample features be 'non-negative'
* NMF achieves its interpretability by decomposing samples as sums of their parts
    * for example: decomposes documents as combinations of common themes
    * images  as comginations of common patterns
    

```python
# Import NMF
from sklearn.decomposition import NMF

# Create an NMF instance: model
model = NMF(n_components=6)

# Fit the model to articles
model.fit(articles)

# Transform the articles: nmf_features
nmf_features = model.transform(articles)

# Print the NMF features
print(nmf_features.round(2))
```

# Building recommender systmes using NMF 

* apply nmf to the word frequency array of the articles
* then use the resulting NMF features
* the NMF features describe the topic mixture of an article
* similar articles will have similar nmf features
* but how can two articles be compared using nmf features????
    * similar articles have similar topics
    * but that doesnt mean the nmf features will be exactly the same
    * one version might use very direct language while another uses meaningless chatter
    * meaningless chatter reduces the values of the nmf features representing the topics
        * however, all these versions lie on a single line
        * thus, article 1 will be on a line with all its version and article 2 will be on its own line with all its versions
        * ENTER TRIGONOMETRY
    * we can use cosine to measure the angle between the two articles and compare how similar the articles are
        * higher values indicate a greater similarity
       

```python
from sklearn.decomposition import NMF
nmf = NMF(n_components=6)
nmf_features = nmf.fit_transform(articles)
```
